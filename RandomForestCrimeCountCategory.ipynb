# Adding Income & median price

```python
def add_income():
    cv=df[["LSOAcode","Mean_income"]].drop_duplicates().set_index("LSOAcode")
    support= cv.to_dict()["Mean_income"]   #dictionary with the key:lsoa code , value:mean income
    
    def add_dataset(x):
        try:
            return support[x]
        except KeyError:
            pass
    final["mean_income"]= final["LSOAcode"].apply(add_dataset)
add_income() 
    
    
```

```python
def add_median_price():
    cv=df[["LSOAcode","Median_price"]].drop_duplicates().set_index("LSOAcode")
    support= cv.to_dict()["Median_price"]   #dictionary with the key:lsoa code , value:mean income
    
    def add_dataset(x):
        try:
            return support[x]
        except KeyError:
            pass
    final["Median_price"]= final["LSOAcode"].apply(add_dataset)
add_median_price()
    
    #Median_price
```
