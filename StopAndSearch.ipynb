```python
labelencoder_Type = LabelEncoder()
labelencoder_Gender = LabelEncoder()
labelencoder_Ethnicity = LabelEncoder()
labelencoder_Agerange = LabelEncoder()
labelencoder_Legislation = LabelEncoder()
labelencoder_Objectofsearch = LabelEncoder()
labelencoder_Outcome = LabelEncoder()

min_max_scaler = MinMaxScaler()
#df[["Median_price", "Mean_income"]] = min_max_scaler.fit_transform(df[["Median_price", "Mean_income"]])

df['Type'] = labelencoder_Type.fit_transform(df['Type'])
df['Gender'] = labelencoder_Gender.fit_transform(df['Gender'])
df['Officer-definedethnicity'] = labelencoder_Ethnicity.fit_transform(df['Officer-definedethnicity'])
df["Legislation"] = labelencoder_Legislation.fit_transform(df['Legislation'])
df["Objectofsearch"] = labelencoder_Objectofsearch.fit_transform(df['Objectofsearch'])
df["Agerange"] = labelencoder_Agerange.fit_transform(df['Agerange'])
df_y["Outcome"] = labelencoder_Outcome.fit_transform(df_y['Outcome'])



df
```

```python

# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(df, df_y, train_size=0.8, test_size=0.2, random_state=0)
from sklearn.model_selection import GridSearchCV
```

```python

#################################################################
# Pipeline
#################################################################
params = {
    'max_depth': 6,
    'objective': 'multi:softprob',  # error evaluation for multiclass training
    'num_class': 3,
    'n_gpus': 0
}
pipe_xgb = Pipeline([
    ('clf', xgb.XGBClassifier(**params))
    ])

parameters_xgb = {
        'clf__n_estimators':[30,40], 
        'clf__criterion':['entropy'], 
        'clf__min_samples_split':[15,20], 
        'clf__min_samples_leaf':[3,4]
    }

grid_xgb = GridSearchCV(pipe_xgb,
    param_grid=parameters_xgb,
    scoring='accuracy',
    cv=5,
    refit=True, n_jobs = -1)

setattr(xgb.XGBClassifier, 'verbosity', 2)
```

```python

import time

#################################################################
# Modeling
#################################################################

grid_xgb.fit(X_train, y_train)
```

```python

#Calculate the score once and use when needed
acc = grid_xgb.score(X_valid,y_valid)
```

```python

print("Best params                        : %s" % grid_xgb.best_params_)
print("Best training data accuracy        : %s" % grid_xgb.best_score_)    
print("Best validation data accuracy (*)  : %s" % acc)
#print("Modeling time                      : %s" % time.strftime("%H:%M:%S", time.gmtime(time.time() - start_time)))
```
